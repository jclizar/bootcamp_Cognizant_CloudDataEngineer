```python
'''LIMPEZA:

ocorrencia_uf
    ** 
ocorrencia_aerodromo
    ###!
    ####
    ****
    *****
ocorrencia_hora
    NULL
'''
```

```python
#NA: not avaiable
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
df.head()
```

```python
#alterar todo o dataframe
df.replace(['**','###!','####','****','*****','NULL'], pd.NA, inplace=True)
```

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
# completa os dados não informados, por exemplo 10
df.fillna(10, inplace=True)
```
